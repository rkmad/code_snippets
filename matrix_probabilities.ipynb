Italian football stats

```python
# pip install -U ipykernel
# pip install -U seaborn
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("stats_teams_1.csv")
```

```python
df = df[(df["Anno"]=="2022-23") | (df["Anno"]=="2021-22") |  (df["Anno"]=="2020-21")]
df["Draw %"]= df["Draw %"].str.replace("%","")
df["Casa Draws"]= df["Casa Draws"].str.replace("%","")
df["Ospite Draws"]= df["Ospite Draws"].str.replace("%","")
```

```python
df_draws= df.groupby("Squadra")[["MP", "Draws"]].sum()
```

```python
df_draws["draw_freq"]= df_draws["Draws"]/df_draws["MP"]
```

```python
df_draws["draw_freq"]
```

```python
vert_teams= pd.DataFrame(df_draws["draw_freq"])
horz_teams= pd.DataFrame(df_draws["draw_freq"])
df_matrix= vert_teams.dot(horz_teams.T)
```

```python
mask= np.triu(df_matrix)
plt.figure(figsize=(12, 12))
sns.heatmap(df_matrix, annot=True, mask=mask,fmt=".3f")
plt.show()
```
